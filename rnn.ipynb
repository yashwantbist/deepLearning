RNN = Neural network that remembers the past

A Recurrent Neural Network (RNN) is designed for sequential data like text, speech, or time-series. Unlike regular neural networks, RNNs pass information from one step to the next, allowing them to remember earlier inputs when processing new ones.

Data Type	Example<br>
<br>Text	Sentence meaning<br>
<br>Speech	Audio recognition<br>
<br>Stock prices	Past trends<br>
<br>Weather	   Yesterday affects today<br>
<br>Chatbots	Conversation context<br>

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load data (variable-length sequences)
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=10000)

# ✅ Pad sequences to same length
maxlen = 200
X_train = pad_sequences(X_train, maxlen=maxlen, padding="post", truncating="post")
X_test  = pad_sequences(X_test,  maxlen=maxlen, padding="post", truncating="post")

# ✅ Make sure dtype is integer (Embedding needs ints)
X_train = X_train.astype("int32")
X_test  = X_test.astype("int32")

# Build model (now input length is fixed)
model = Sequential([
    Embedding(input_dim=10000, output_dim=64, input_length=maxlen),
    SimpleRNN(64),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=3, batch_size=64, validation_data=(X_test, y_test))
```
